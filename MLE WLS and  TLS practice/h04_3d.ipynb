```python
import scipy.io as sio
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
```

```python
sigmal = []
sigmal = [[[1,0],[0,1]],[[1, 0.25],[0.25, 1]],[[1, 0.9],[0.9, 1]],[[1, -0.25],[-0.25, 1]],[[1, -0.9],[-0.9, 1]],[[1, 0.1],[0.1, 1]]]
sampleNuml = [5, 50, 500]
```

```python
delta = 0.025
xg = np.arange(0, 2, delta)
yg = np.arange(0, 2, delta)
X, Y = np.meshgrid(xg, yg)
testSampleNum = 100
for sig in sigmal:
    for sampleNum in sampleNuml:
        z = np.random.normal(0,1,sampleNum)
        x1 = np.random.normal(0,5,sampleNum)
        x2 = np.random.normal(0,5,sampleNum)
        x = np.concatenate((x1,x2),axis = 0)
        x.shape = (2,sampleNum)
        y = x1 + x2 + z
        mean = np.dot(np.linalg.inv(np.dot(x,x.T) + np.linalg.inv(sig)),np.dot(x,y))
        var = np.linalg.inv(np.dot(x,x.T)+ np.linalg.inv(sig))
        Z = mlab.bivariate_normal(X, Y, np.sqrt(var[0][0]), np.sqrt(var[1][1]), mean[0], mean[1],var[0][1])
        plt.figure()
        plt.contour(X, Y, Z)
        plt.title('sigma^2 = {} n = {}'.format(sig,sampleNum))
        plt.scatter(1,1,marker = 'D',c = 'red')
plt.show()
```
