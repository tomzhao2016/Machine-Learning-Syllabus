```python
import scipy.io as sio
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
```

```python
sigmal = [[[1,0],[0,1]],[[1, 0.25],[0.25, 1]],[[1, 0.9],[0.9, 1]],[[1, -0.25],[-0.25, 1]],[[1, -0.9],[-0.9, 1]],[[1, 0.1],[0.1, 1]]]
sampleNuml = range(5,200,5)
delta = 0.025
xg = np.arange(0, 2, delta)
yg = np.arange(0, 2, delta)
X, Y = np.meshgrid(xg, yg)
testSampleNum = 100
x1_test = np.random.normal(0,5,testSampleNum)
x2_test = np.random.normal(0,5,testSampleNum)
x_test = np.concatenate((x1_test,x2_test),axis = 0)
z_test = np.random.normal(0,1,testSampleNum)
x_test.shape = (2,testSampleNum)
y_test = x1_test + x2_test + z_test
avg = 20

for sig in sigmal:
    error = []
    theo_error = []
    mean = 0
    for sampleNum in sampleNuml:
        for i in range(avg):
            z = np.random.normal(0,1,sampleNum)
            x1 = np.random.normal(0,5,sampleNum)
            x2 = np.random.normal(0,5,sampleNum)
            x = np.concatenate((x1,x2),axis = 0)
            x.shape = (2,sampleNum)
            y = x1 + x2 + z
            mean += np.dot(np.linalg.inv(np.dot(x,x.T) + np.linalg.inv(sig)),np.dot(x,y))
            #error.append(np.linalg.norm(y_test - (mean[0]*x1_test + mean[1]*x2_test),2))
            var = np.linalg.inv(np.dot(x,x.T)+ np.linalg.inv(sig))
        mean = mean/avg
        error.append(np.linalg.norm(y_test - mean[0]*x1_test - mean[1]*x2_test,2)**2/sampleNum)
        theo_error.append(5*(mean[0]-1)**2 + 5*(mean[1]-1)**2 + 1)
    plt.plot(sampleNuml,error)
    plt.plot(sampleNuml,theo_error)
    plt.title('sigma^2 = {}'.format(sig))
    plt.show()
```
